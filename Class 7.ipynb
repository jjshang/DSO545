```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Question 1 -  Density plots

data = pd.read_csv("mpg.csv")
```

```python
data.head()
```

```python
plt.figure()

# Use Seaborn to create kernel density plot (kdeplot)
sns.kdeplot(data.loc[data.cyl ==4, "hwy"], 
            shade = True,
           color = "green",
           label = "Cyl = 4",
           alpha = 0.4)

sns.kdeplot(data.loc[data.cyl ==5, "hwy"], 
            shade = True,
           color = "deeppink",
           label = "Cyl = 5",
           alpha = 0.4)

sns.kdeplot(data.loc[data.cyl ==6, "hwy"], 
            shade = True,
           color = "dodgerblue",
           label = "Cyl = 6",
           alpha = 0.4)

sns.kdeplot(data.loc[data.cyl ==8, "hwy"], 
            shade = True,
           color = "orange",
           label = "Cyl = 8",
           alpha = 0.4)

plt.show()
```

```python
# Question 2 - create a 2x3 grid

plt.figure()

## fig, ax = figure for graph. Each subplot is assigned a number and can be called 
## plt.subplots calls subplots. Specify the dimension in bracket
fig, ax = plt.subplots(2,3)

plt.show()
```

```python
plt.subplot(2,3,1)
plt.subplot(2,3,2)
plt.subplot(2,3,4)
```

```python
plt.figure(figsize = (10,5))

plt.subplots_adjust(hspace = 0.5, wspace = 0.6) ## hspace = height space; space between rows, wspace = width space; space between columns
 
for i in range(1,7):
    plt.subplot(2,3,i)
    plt.text(0.5, 0.5, str((2,3,i)), ha = "center") ## ha: horizontal alignment

plt.show()
```

```python
# Question 3 - create a 2x3 graph with sin and cos plots

plt.figure()

fig, ax = plt.subplots(2,3, sharex = 'col', sharey = 'row') ## sharex / sharey = share x/y-axis for all subplots

x = np.linspace(0,50) #linspace = linear space equally spaced
ax[0,0].plot(np.sin(x))
ax[1,2].plot(np.cos(x), color = 'red')
ax[1,1].plot(np.tan(x), color = 'green')


plt.show()
```

```python
# Question 4

fig = plt.figure()

grid = plt.GridSpec(2,3,wspace = 0.4, hspace = 0.5)

fig.add_subplot(grid[0,0])
fig.add_subplot(grid[0,1:3])
fig.add_subplot(grid[1,0:2])
fig.add_subplot(grid[1,2])

plt.show()
```

```python
# Question 5 - Marginal Boxplots

fig = plt.figure(figsize = (10,5))

grid = plt.GridSpec(4,4,wspace = 0.5, hspace = 1)

ax_main = fig.add_subplot(grid[0:-1,0:-1]) ## use -1 in case you decide to change spec to avoid changing these codes
ax_right = fig.add_subplot(grid[0:-1,-1])
ax_bottom = fig.add_subplot(grid[3,0:3])

## Plot main graph (scatter plot)
ax_main.scatter('displ', 'hwy', data = data)

## Plot the right boxplot
sns.boxplot(data.hwy, ax = ax_right, orient = 'v') ## orient='v': orientation is vertical
ax_right.set(xticks = [], yticks = [], ylabel = "") ## more than one subplot, have to set axis details in brackets

## Plot the bottom boxplot
sns.boxplot(data.displ, ax = ax_bottom)
ax_bottom.set(xticks = [], yticks = [], xlabel = "")

## Set title, x-label, y-label for main graph
ax_main.set(title = "Scatterplot with Boxplots \ndispl vs hwy mileage",
           xlabel = "Displacement",
           ylabel = "Hwy Mileage")

plt.show()
```

```python
# Question 6

data = pd.read_csv("economics.csv")
```

```python
data.head()
```

```python
fig = plt.figure(figsize = (12,5))

plt.plot(data.date, data.psavert, color = 'tab:red')
ax = plt.gca()

## Set the main y-axis (left)
ax.tick_params(axis = 'y', labelcolor = 'tab:red')
ax.set_ylabel("Personal Savings Rate", color = 'tab:red', fontsize = 14)

## Set the secondary y-axis (right)
ax_right = ax.twinx() ### create a twin y-axis that share the same x-axis
ax_right.plot(data.date, data.unemploy, color = 'tab:blue')
ax_right.tick_params(axis = 'y', labelcolor = 'tab:blue')
ax_right.set_ylabel("Unemployed (1000's)", color = 'tab:blue', fontsize = 14)

## Set the x-axis
ax.set_xticks(np.arange(0,len(data.date),60))
ax.set_xlabel("Date", fontsize = 14)

plt.title("Personal Savings Rate vs. Unemployment", fontsize = 20)

plt.show()
```
